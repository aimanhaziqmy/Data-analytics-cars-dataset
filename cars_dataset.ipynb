### 1) Data Cleaning

1. Find all null value in the dataset, fill with the mean of the column
2. If make is null, remove the row


```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
car = car[~car['Make'].isnull()]
```

### 3. Filtering

Show all the records where Origin is Asia or Europe

```python
car.head(2)
```

### 4. Removing unwanted records

Remove all records that have 400 above in weight

```python
car = car[~(car['Weight'] > 4000)]
```

### 5. Applying function to a column

Increase all the values of 'MPG_City' column by 3.

```python
car['MPG_City'] = car['MPG_City'].apply(lambda x:x+3)
```
